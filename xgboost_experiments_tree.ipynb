```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_final.drop(columns=['text', 'norm_rating']).to_numpy(), df_final.norm_rating.to_numpy(), random_state=42, test_size=0.2)
```

```python
y_train = list(y_train)
y_test = list(y_test)
```

```python
import os
import numpy as np
import xgboost as xgb
```

```python
params = {
        'eta': 0.2,
        'max_depth': 3,
        'min_child_weight': 2,
        'gamma': 0,
        'subsample': 0.9,
        'colsample_bytree': 0.8,
        'reg_alpha': 0.00001,
        'seed': 42,
        'objective': 'multi:softmax',
        'num_class': 3,
    }
```

```python
import scipy.sparse
```

```python
csr_train = scipy.sparse.csr_matrix(x_train)
dtrain = xgb.DMatrix(csr_train, label=y_train)

csr_test = scipy.sparse.csr_matrix(x_test)
dtest = xgb.DMatrix(csr_test, label=y_test)
```

```python
csr = scipy.sparse.csr_matrix(x_test)
dtest2 = xgb.DMatrix(csr)
```

```python
watchlist = [(dtest, 'eval'), (dtrain, 'train')]
num_round = 20
bst = xgb.train(params, dtrain, num_round, watchlist)
```
